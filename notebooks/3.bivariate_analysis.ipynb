**Bivariate Analysis**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/productivity_raw.csv')
```

*Study Hours vs. Music*

```python
m_sh = sns.scatterplot(x='Study_Hours', y='Music', data=df)
m_sh.set_title('Study Hours vs. Music')
plt.show()
```

*Temperature vs. Study Hours*

```python
t_sh = sns.scatterplot(x='Temperature', y='Study_Hours',hue='Mood', data=df)
t_sh.set_title('Temperature vs. Study Hours')
plt.show()
```

```python
sns.pairplot(df, vars=['Temperature', 'Study_Hours'], hue='Mood')
plt.show()
```
