**Bivariate Analysis**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/productivity_raw.csv')
```

*Music vs. Study Hours*

```python
m_sh = sns.scatterplot(x='Music', y='Study_Hours', data=df)
m_sh.set_title('Music vs. Study Hours')
plt.show()
```

*Music vs. Mood*

```python
ms_m = sns.scatterplot(x='Music', y='Mood_Score', data=df)
ms_m.set_title('Music vs. Mood')
plt.show()
```

*Mood vs. Study Hours*

```python
ms_sh = sns.scatterplot(x='Mood_Score', y='Study_Hours', data=df)
ms_sh.set_title('Mood vs. Study Hours')
plt.show()
```

*Temperature vs. Study Hours*

```python
t_sh = sns.scatterplot(x='Temperature', y='Study_Hours', data=df)
t_sh.set_title('Temperature vs. Study Hours')
plt.show()
```

*Mood vs. Temperature*

```python
t_m = sns.scatterplot(x='Mood_Score', y='Temperature', data=df)
t_m.set_title('Temperature vs. Mood')
plt.show()
```

*Temperature vs. Study Hours vs. Mood Score*

```python
sns.pairplot(df, vars=['Temperature', 'Study_Hours', 'Mood_Score'], hue='Music')
plt.show()
```
