**Insights Summary**

```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy.ma.core import count
```

```python
df = pd.read_csv('../data/productivity_raw.csv')
```

*Top 3 most productivity music genres (study hours > 3hrs)*

```python
ax = df.query('Music != "Other"') \
    .groupby('Music')['Study_Hours'] \
    .mean() \
    .reset_index() \
    .query('Study_Hours > 3') \
    .sort_values('Study_Hours') \
    .set_index('Music')['Study_Hours'] \
    .head(3) \
    .plot(kind='barh', title='Music Genres with Average Study Hours > 3h')
ax.set_xlabel('Average Study Hours')
plt.show()
```
