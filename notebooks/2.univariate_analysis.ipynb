**Univariate Analysis**

```python
from importlib.metadata import Distribution

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/productivity_raw.csv')
```

*Weather Distribution*

```python
wd = df['Weather'].value_counts() \
    .plot(kind='bar', title='Weather Distribution')
wd.set_xlabel('Weather')
wd.set_ylabel('Count')
plt.show()
```

*Music Distribution*

```python
md = df['Music'].value_counts() \
    .plot(kind='bar', title='Music Distribution')
md.set_xlabel('Music')
md.set_ylabel('Count')
plt.show()
```

*Mood Distribution*

```python
ms = df['Mood_Score'].plot(kind='hist', title='Mood Distribution')
ms.set_xlabel('Mood Score')
plt.show()
```

*Temperature Distribution*

```python
td = df['Temperature'].plot(kind='hist', title='Temperature Distribution')
td.set_xlabel('Temperature')
plt.show()
```

*Study Hours Distribution*

```python
sh = df['Study_Hours'].plot(kind='hist', title='Study Hours Distribution')
sh.set_xlabel('Study Hours')
plt.show()
```
